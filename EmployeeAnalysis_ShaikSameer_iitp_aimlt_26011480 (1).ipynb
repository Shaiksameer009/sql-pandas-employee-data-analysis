This project demonstrates how to create a synthetic employee database using SQLite and analyze it using SQL and Pandas.


Task 1: Create Database

```python
import os
if os.path.exists("employees.db"):
    os.remove("employees.db")
```

```python
import sqlite3
import pandas as pd
import random
```

```python
#Create SQLite Database
conn = sqlite3.connect("employees.db")
cursor = conn.cursor()
print("Database created successfully")
```

```python
#Create Employees Table
cursor.execute("""
CREATE TABLE IF NOT EXISTS employees (
    emp_id INTEGER PRIMARY KEY,
    name TEXT,
    department TEXT,
    salary REAL,
    years_experience INTEGER,
    performance_score REAL
)
""")
print("Employees table created successfully")
```

```python
#Generate Synthetic Employee Data
departments = ["Engineering", "Sales", "Marketing", "HR", "Finance"]
employees = []
for i in range(1, 41):
    name = f"Employee_{i}"
    department = random.choice(departments)
    salary = random.randint(50000, 150000)
    years_exp = random.randint(1, 15)
    performance = round(random.uniform(1.0, 5.0), 2)
    employees.append((i, name, department, salary, years_exp, performance))
```

```python
#Insert Employee Records
cursor.executemany("""
INSERT INTO employees VALUES (?, ?, ?, ?, ?, ?)
""", employees)
conn.commit()
print("40 employee records inserted successfully")
```

Task 2 – SQL Queries

Query 1: High Performing Employees

```python
#This query finds employees with performance score greater than or equal to 4.0 and at least 3 years of experience. The results are sorted by performance score in descending order and limited to the top 15 employees.
query1 = """
SELECT name, department, salary, performance_score
FROM employees
WHERE performance_score >= 4.0
AND years_experience >= 3
ORDER BY performance_score DESC
LIMIT 15
"""

df_query1 = pd.read_sql(query1, conn)

df_query1
```

Query 2: Employees in Engineering or Sales with Salary Between 70k and 110k

```python
query2 = """
SELECT *
FROM employees
WHERE salary BETWEEN 70000 AND 110000
AND department IN ('Engineering', 'Sales')
ORDER BY department ASC, salary DESC
"""

df_query2 = pd.read_sql(query2, conn)

df_query2
```

Query 3: Employee Count and Average Salary by Department

```python
#This query calculates the total number of employees and the average salary for each department.  The results are sorted by average salary in descending order.
query3 = """
SELECT department,
COUNT(*) AS employee_count,
AVG(salary) AS avg_salary
FROM employees
GROUP BY department
ORDER BY avg_salary DESC
"""
df_query3 = pd.read_sql(query3, conn)
df_query3
```

Task 3 – Pandas Analysis

```python
#Load Employee Data into Pandas
df = pd.read_sql("SELECT * FROM employees", conn)
df.head()
```

```python
#Average Performance Score by Department
avg_performance = df.groupby("department")["performance_score"].mean()
avg_performance
```

```python
### Replicating Query 1 Using Pandas
pandas_query1 = df.loc[
    (df["performance_score"] >= 4.0) &
    (df["years_experience"] >= 3),
    ["name", "department", "salary", "performance_score"]
].sort_values(by="performance_score", ascending=False).head(15)
pandas_query1
```

 #150-200 words comparing SQL vs Pandas:

SQL and Pandas are both powerful tools for data analysis, but they are used in different contexts and have different syntax styles. SQL is mainly used to query and manage data stored in relational databases. It uses commands such as `SELECT`, `WHERE`, `AND`, and `ORDER BY` to filter, sort, and retrieve data. For example, SQL uses `=` for comparison and logical operators like `AND` to combine conditions.

Pandas, on the other hand, is a Python library used for in-memory data analysis. Instead of SQL commands, Pandas uses Python syntax such as `==` for equality, `&` for combining conditions, and functions like `.sort_values()` for sorting data. Data is stored in DataFrames, which allow flexible data manipulation and analysis.

SQL is best used when working directly with large databases because it can efficiently query data stored on disk without loading everything into memory. Pandas is more suitable for detailed data exploration, transformation, and analysis once the data has been loaded into memory.

For datasets that are too large to fit into memory, SQL is the better option because the database engine processes the data internally and returns only the required results.
